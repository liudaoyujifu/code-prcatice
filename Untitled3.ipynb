Create an array

```python
import numpy

x = numpy.random.uniform(0.0, 5.0, 500)

print(x)
```

Draw a histogram

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 500)

plt.hist(x, 5)
plt.show()
```

Create a big array

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```

A typical normal data distribution

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 100000)

plt.hist(x, 100)
plt.show()
```
