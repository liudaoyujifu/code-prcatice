Linear Regression  （Import scipy and draw the line of Linear Regression:）

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [6,13,9,15,2,12,2,9,5,11,12,5,6]
y = [89,86,87,88,111,87,103,87,105,78,79,85,89]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

Predict Future Values

```python
from scipy import stats

x = [6,13,9,15,2,12,2,9,5,11,12,5,6]
y = [89,86,87,88,111,87,103,87,105,78,79,85,89]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)

print(speed)
```
